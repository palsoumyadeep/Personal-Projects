```python
df_copy = df.copy()
```

```python
df_copy = df_copy[['Age', 'Sex',
       'Category_Name', 'Surgery_Code','Preauth_Date', 'Preauth_Amount',
       'Claim_Date', 'Claim_Amount','Hospital_Type','Surgery_Date',
       'Discharge_Date','Mortality_Yes/No','Hospital_District']]
```

```python
df_copy.columns
```

```python
date_columns = ['Preauth_Date',"Claim_Date", 'Surgery_Date', 'Discharge_Date']
for col in date_columns:
  df_copy[col] = pd.to_datetime(df[col], format='%d-%m-%y')
```

```python
df_copy["pre - claim"] = -(df_copy['Preauth_Date'] - df_copy["Claim_Date"]).dt.days
df_copy["surg - claim"] = -(df_copy['Surgery_Date'] - df_copy["Claim_Date"]).dt.days
df_copy["pre - surg"] = -(df_copy['Preauth_Date'] - df_copy["Surgery_Date"]).dt.days
df_copy["claim - dicharge"] = -(df_copy['Claim_Date'] - df_copy["Discharge_Date"]).dt.days
df_copy["surg - discharge"] = -(df_copy['Surgery_Date'] - df_copy["Discharge_Date"]).dt.days
df_copy["total_days"] = -(df_copy['Preauth_Date'] - df_copy["Discharge_Date"]).dt.days
```

```python
df_copy.drop(date_columns, axis = 1 , inplace = True)
```

```python
df_copy.head()
```

```python
cat_cols = ['Sex', 'Category_Name','Surgery_Code',
       'Hospital_Type', 'Hospital_District',
      'Mortality_Yes/No']
```

```python
def get_dummy_columns(dataframe, categorical_columns):
    # Create dummy columns for the categorical variables
    dataframe = pd.get_dummies(dataframe, columns=categorical_columns, drop_first=True)

    return dataframe
```

```python
df_copy.columns
```

```python
model_1 = ['Age', 'Category_Name', 'Surgery_Code', 'Preauth_Amount',
       'Claim_Amount', 'Hospital_Type', 'Mortality_Yes/No',
       'Hospital_District', 'pre - claim', 'surg - claim', 'pre - surg',
       'claim - dicharge', 'surg - discharge', 'total_days']

model_2 = ['Age', 'Sex', 'Category_Name', 'Surgery_Code', 'Preauth_Amount',
       'Claim_Amount']

model_t = ['Age', 'Sex', 'Category_Name', 'Surgery_Code', 'Preauth_Amount',
       'Claim_Amount', 'Hospital_Type', 'Mortality_Yes/No',
       'Hospital_District', 'pre - claim', 'surg - claim', 'pre - surg',
       'claim - dicharge', 'surg - discharge', 'total_days']
```

```python
def find_intersection(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    intersection = set1.intersection(set2)
    return list(intersection)
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

def extended_isolation_forest(dataframe,columns, n_estimators=500, contamination=0.1):
    dataframe = dataframe[columns]
    categorical_cols = find_intersection(columns,cat_cols)

    dataframe = get_dummy_columns(dataframe,categorical_cols)

    # # Assuming the dataframe has features in columns 0 to n-1, where n is the number of features
    # X = dataframe.iloc[:, :-1].values

    # Create an instance of Extended Isolation Forest
    eif = IsolationForest(n_estimators=n_estimators, contamination=contamination)

    # Fit the model and obtain the anomaly score
    anomaly_scores = eif.fit_predict(dataframe)

    return anomaly_scores
```

```python
df_copy["Score_1"] = extended_isolation_forest(df_copy,model_1)
df_copy["Score_2"] = extended_isolation_forest(df_copy,model_2)
df_copy["Score_t"] = extended_isolation_forest(df_copy,model_t)
```
