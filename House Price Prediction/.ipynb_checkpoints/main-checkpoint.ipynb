```python
data.drop(['BHK_OR_RK','POSTED_BY'], axis = 1, inplace = True)
```

```python
#concatination of dummy variables to original data frame
merged_df =  pd.concat([dummies_posted_by,dummies_bhk_rk,data], axis = 1)
merged_df
```

```python
merged_df.drop(["BUILDER","BHK"])
            
```

```python
#dropping outliers using z-score
index_to_drop = []

for col in merged_df.columns : 
    z_scores = stats.zscore(merged_df[col])
    for index in range(len(z_scores)):
        if abs(z_scores[index]) > 3:
            index_to_drop.append(index)
            
            
index_to_drop = sorted(list(set(index_to_drop)))            
final_df = merged_df.drop(index_to_drop)
final_df.reset_index(drop=True, inplace=True)
```

```python

merged_df.shape
```

```python
final_df.shape
```

```python
y = merged_df.iloc[:, -1].values # price column
X = merged_df.iloc[:, :-1].values
```

```python
#train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred_mlr = model.predict(X_test)
r2_mlr = r2_score(y_test, y_pred_mlr)
mae_mlr = mean_absolute_error(y_test, y_pred_mlr)
print("R2 score for multi linear Regression:  {}".format(r2_mlr))
print("Mean Absolute Error for multi linear Regression:  {}".format(mae_mlr))
```
